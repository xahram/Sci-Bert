```python
import pandas as pd

tweet_df = pd.read_excel('/gdrive/My Drive/Master_dataset/U-CRISE TRUE AND FALSE CLAIMS(1).xlsx')  


tweet_df.head()
# tweet_df.dropna()
```

```python
import re


# function to rmeove digits and numbers from papers 
def tweet_preprocessing(tweet):      
    
    #####################################


    tweet = re.sub(r"[~.,%/:;?_&+*=!-]", ' ',tweet)
    
    tweet = re.sub(r'[^a-z]', ' ', tweet)
    
    tweet = re.sub(r'\s{2,}', ' ', tweet) 
                                         
    
    

    return tweet




tweet_df["TWEET CONTENT"] = tweet_df["TWEET CONTENT"].dropna(how='any')
tweet_df["TWEET CONTENT"]
```

```python

import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize
nltk.download('punkt')
# Load all necessary libraries

import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk import word_tokenize


nltk.download("punkt")
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()


stop_words = set(stopwords.words("english"))


tweet_df['processed_tweet'] = tweet_df["TWEET CONTENT"].dropna().apply(tweet_preprocessing)
processed_tweets = tweet_df["processed_tweet"].tolist()
print(type(processed_tweets))
print(len(processed_tweets))


processed_tweets = [str(i) for i in processed_tweets]



def vocabulary(C):
    print(f'helllowowo {C}')
    word_tokens = word_tokenize(C)
    v = [lemmatizer.lemmatize(w) for w in word_tokens if w not in stop_words and w != 'nan']
    return v


# processd_tweets_corpus = vocabulary(processd_tweets_corpus[:30])
processd_tweets_corpus = " ".join(processed_tweets)
processd_tweets_corpus
v = vocabulary(processd_tweets_corpus)
v
```
